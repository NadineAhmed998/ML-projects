```python
def polynomial_train_cv(degree):
  "Creates a polynomial regression model for the given degree"
  
  poly_features = PolynomialFeatures(degree=degree)
  
  # transforms the existing features to higher degree features.
  X_train_poly = poly_features.fit_transform(X_train)
  
  # fit the transformed features to Linear Regression
  poly_model = LinearRegression()
  poly_model.fit(X_train_poly, y_train)
  
  # predicting on training data-set
  y_train_predicted = poly_model.predict(X_train_poly)
  
  # predicting on cv data-set
  y_cv_predict = poly_model.predict(poly_features.fit_transform(X_cv))
  
  # evaluating the model on training dataset
  rmse_train = np.sqrt(mean_squared_error(y_train, y_train_predicted))
  r2_train = r2_score(y_train, y_train_predicted)
  
  # evaluating the model on cv dataset
  rmse_cv = np.sqrt(mean_squared_error(y_cv, y_cv_predict))
  r2_cv = r2_score(y_cv, y_cv_predict)
    

  print("The model performance for the training set for degree " + str(degree))
  print("-------------------------------------------")
  print("RMSE of training set is {}".format(rmse_train))
  print("R2 score of training set is {}".format(r2_train))
  
  print("\n")
  
  print("The model performance for the cv set")
  print("-------------------------------------------")
  print("RMSE of cv set is {}".format(rmse_cv))
  print("R2 score of cv set is {}".format(r2_cv))
  print("end end end end end ")  
  return rmse_cv  
```
